```python
df_clean['rooms_per_household'] = df_clean['total_rooms'] / df_clean['households']
df_clean['bedrooms_per_room'] = df_clean['total_bedrooms'] / df_clean['total_rooms']
df_clean['population_per_household'] = df_clean['population'] / df_clean['households']
```

```python
df_clean['housing_median_age'] = df_clean['housing_median_age'].astype(int)
```

```python
Q1 = df_clean['median_house_value'].quantile(0.25)
Q3 = df_clean['median_house_value'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df_clean = df_clean[
    (df_clean['median_house_value'] >= lower) &
    (df_clean['median_house_value'] <= upper)
]
```

```python
df_clean['ocean_proximity'].value_counts()
```

```python
df_clean=df_clean.dropna()
```

```python
df_clean.columns
```

```python
y = df_clean['median_house_value']
```

```python
y
```

```python
y.shape
```

```python
X=df_clean[[ 'median_income',
        'housing_median_age',
        'total_rooms',
        'total_bedrooms',
        'population',
        'households',
        'longitude',
        'latitude',
        'ocean_proximity']]
```

```python
X
```

```python
# ONE HOT ENCODING 
X = pd.get_dummies(X, columns=['ocean_proximity'], drop_first=True)
```

```python
X
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train , y_test= train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
print(X_train.head())
print(y_train.head())
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
new_house = [[
    8.3252,   # median_income
    41,       # housing_median_age
    880,      # total_rooms
    140,      # total_bedrooms
    322,      # population
    126,      # households
    -122.23,  # longitude
    37.88,    # latitude
    0,        # ocean_proximity_INLAND
    0,        # ocean_proximity_ISLAND
    1,        # ocean_proximity_NEAR BAY
    0         # ocean_proximity_NEAR OCEAN
]]
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
X_train_scaled = pd.DataFrame(
    X_train_scaled,
    columns=X_train.columns,
    index=X_train.index
)

X_test_scaled = pd.DataFrame(
    X_test_scaled,
    columns=X_test.columns,
    index=X_test.index
)
```

```python
X_train_scaled 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train_scaled, y_train)
```

```python
# make predicted 
y_train_pred = lr_model.predict(X_train_scaled)
y_test_pred  = lr_model.predict(X_test_scaled)
```

```python
y_train_pred
```

```python
y_test_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mean_squ= mean_squared_error(y_test, y_test_pred)

rmse= mean_squ**0.5
r2=r2_score(y_test, y_test_pred)

```

```python
rmse
```

```python
r2
```

```python
# Make Predictions

y_pred = lr_model.predict(X_test_scaled)
```

```python
import pandas as pd

results = pd.DataFrame({
    "Actual Price": y_test,
    "Predicted Price": y_pred
})

print(results.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted House Prices")
plt.show()
```
